```python
movies = pd.read_csv("/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv")
cred = pd.read_csv("/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv")
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
import ast
def convert(dict_genre):
    genre = []
    for i in ast.literal_eval(dict_genre):
        genre.append(i['name']) 
    return genre 
```

```python
movies['genres']=movies['genres'].apply(convert)
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head(10)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies_new = movies[['movie_id','title','tags']]
```

```python
movies_new.head()
```

```python
movies_new['tags'] = movies_new['tags'].apply(lambda x: " ".join(x))
```

```python
movies_new['tags'] = movies_new['tags'].apply(lambda x: x.lower())
```

```python
embedding_vec_2 = model.encode(movies_new['tags'].tolist())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(embedding_vec_2)
```

```python
def recommend(movie_title, top_n=8):
    movie_index = movies_new[movies_new['title'] == movie_title].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:top_n+1]
    #movie_list = [(1, 0.9), (2, 0.85), (3, 0.4)]
    
    for i in movie_list:
        print(movies_new.iloc[i[0]]['title'])
```

```python
recommend("Batman")
```
